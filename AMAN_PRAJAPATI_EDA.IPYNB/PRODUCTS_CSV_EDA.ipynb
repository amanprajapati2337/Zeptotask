```python
# import all the libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('products.csv')# load the data
print(data.head())
```

```python
print(data.info())
```

```python
print(data.describe())
summary_stats = data.describe()
duplicates = data.duplicated().sum()
```

```python
category_distribution = data['Category'].value_counts()# Distribution of products across categories
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Category', order=category_distribution.index, palette='viridis')
plt.title('Distribution of Products Across Categories', fontsize=16)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.show()

```

```python
# Price distribution
plt.figure(figsize=(10, 6))
sns.histplot(data['Price'], kde=True, bins=15, color='skyblue')
plt.title('Price Distribution', fontsize=16)
plt.xlabel('Price', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()

```

```python
# Boxplot for outlier detection
plt.figure(figsize=(10, 6))
sns.boxplot(x=data['Price'], color='orange')
plt.title('Boxplot of Prices', fontsize=16)
plt.xlabel('Price', fontsize=12)
plt.show()

```

```python
# Most/Least expensive categories
category_price_stats = data.groupby('Category')['Price'].agg(['mean', 'min', 'max']).sort_values(by='mean', ascending=False)

summary_stats, category_distribution, category_price_stats
```
